```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
df["Airline"].value_counts().plot(kind="bar", color="#B0E0E6")
plt.xlabel("Airline")
plt.ylabel("Number of Flights")
plt.title("Number of Flights by Airline")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

```python
status_counts = df["Status"].value_counts()
status_counts.plot(kind="pie", autopct="%1.1f%%", colors=["#AEC6CF", "#D8BFD8", "#BEBEBE"])
plt.title("Flight Status Distribution")
plt.ylabel("")
plt.show()
```

```python
df["Route"] = df["Origin"] + " â†’ " + df["Destination"]
top_routes = df["Route"].value_counts().head(10)

top_routes.plot(kind="barh", color="#C3B1E1")
plt.xlabel("Number of Flights")
plt.ylabel("Route")
plt.title("Top 10 Most Common Flight Routes")
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df['Fare'].hist(ax=axes[0], bins=20, color='blue', alpha=0.7)
axes[0].set_title('Fare Distribution')

df['Cabin_Baggage_KG'].hist(ax=axes[1], bins=20, color='green', alpha=0.7)
axes[1].set_title('Cabin Baggage Weight')

df['Checkin_Baggage_KG'].hist(ax=axes[2], bins=20, color='red', alpha=0.7)
axes[2].set_title('Checkin Baggage Weight')

plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Month', hue='Year', palette='viridis')
plt.title("Monthly Passenger Traffic")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

plt.figure(figsize=(12, 6))
sns.lineplot(data=df.groupby(['Year', 'Month']).size().reset_index(name='Count'), x='Month', y='Count', hue='Year', marker='o')
plt.title("Seasonal Trends in Passenger Traffic")
plt.xlabel("Month")
plt.ylabel("Passenger Count")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid()
plt.show()
```
