1. Handling Missing Values
Missing values occur when data is incomplete or unavailable. They can impact data analysis and machine learning models, leading to incorrect results.

Approaches to Handle Missing Values:
Removing Missing Data (Dropping Rows or Columns)

If a row or column has too many missing values, it might be best to remove it.
This is effective when missing data is minimal, but excessive removal may lead to data loss.
Imputation (Filling Missing Values)

Mean, Median, or Mode Imputation:
Numerical values can be replaced with the mean (average), median (middle value), or mode (most frequent value).
Constant Value Imputation:
Categorical values can be filled with placeholders like "Unknown" or "Not Available".
Forward Fill (ffill) and Backward Fill (bfill):
Missing values are filled using the previous (ffill) or next (bfill) available value.
Prediction-Based Imputation

Use machine learning models to predict missing values based on other available data.

```python
# prompt: 1.	Plot the distribution of the following discrete features:
# o	Quantity
# o	Category
# o	Region

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a Pandas DataFrame.
df = pd.read_csv("/content/retail_sales_dataset.csv")

# Rename the columns to more meaningful names, if necessary.
df = df.rename(columns={'Customer ID':'cust_ID'})

# Handle missing values appropriately (drop or impute them).
for column in df.columns:
  if df[column].isnull().any():
    if pd.api.types.is_numeric_dtype(df[column]):
      df[column].fillna(df[column].mean(), inplace=True)
    else:
      df[column].fillna(df[column].mode()[0], inplace=True)

# If there are duplicates, remove them.
df.drop_duplicates(inplace=True)

# Plot the distribution of 'Quantity'
plt.figure(figsize=(8, 6))
sns.histplot(df['Quantity'], kde=True)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of 'Category'
plt.figure(figsize=(10, 6))
sns.countplot(y='Product Category', data=df, order = df['Product Category'].value_counts().index)
plt.title('Distribution of Category')
plt.xlabel('Count')
plt.ylabel('Product Category')
plt.show()

# Plot the distribution of 'Region'
plt.figure(figsize=(8, 6))
sns.countplot(y='Age', data=df, order = df['Age'].value_counts().index)
plt.title('Distribution of Region')
plt.xlabel('Count')
plt.ylabel('Age')
plt.show()
```
